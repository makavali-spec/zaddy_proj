```python
from scipy import stats
# Calculate correlations between spine types and spine densities
spine_types = ['Filopodia', 'Longthin', 'Thin', 'Mushroom', 'Stubby', 'Bifurcated', 'Critical', 'Non-critical']
correlations = df[spine_types].corr()

# Plot correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlations, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Between Spine Types and Spine Densities')
plt.show()

# Calculate effect size (Cohen's d) for spine density differences between Pubertal and Postpubertal stages
pubertal_densities = df[df['Developmental_Stage'] == 'Pubertal']['Total']
postpubertal_densities = df[df['Developmental_Stage'] == 'Postpubertal']['Total']
effect_size = (pubertal_densities.mean() - postpubertal_densities.mean()) / pubertal_densities.std()

print("Effect Size (Cohen's d) for Spine Density between Pubertal and Postpubertal stages:", effect_size)

# Correct for multiple comparisons using the Benjamini-Hochberg procedure
p_values = stats.ttest_ind(pubertal_densities, postpubertal_densities).pvalue
_, corrected_p_values, _, _ = multitest.multipletests(p_values, method='fdr_bh')

# Plot stacked bar plot with significance bars
plt.figure(figsize=(10, 6))
sns.barplot(data=melted_df, x='Developmental_Stage', y='Spine Density', hue='Spine Type', dodge=True)
plt.title('Spine Type Distribution by Developmental Stage and Genotype')
plt.ylabel('Spine Density')
plt.legend(title='Spine Type')

# Add significance bars
for i, p_value in enumerate(corrected_p_values):
    if p_value < 0.05:
        plt.annotate('*', xy=(i % 2, 0), textcoords='offset points', xytext=(0,5), ha='center')

plt.show()

# Compare Wildtype and Knockout groups
wildtype_densities = df[df['Genotype'] == 'Wildtype']['Total']
knockout_densities = df[df['Genotype'] == 'Knockout']['Total']

t_statistic, p_value = stats.ttest_ind(wildtype_densities, knockout_densities)
print("T-test: Wildtype vs Knockout Spine Densities")
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

```python
# Rename columns for clarity
df.rename(columns={"Total": "Total Spine Density (per 10 Âµm)"}, inplace=True)

# Melt the data for plotting
structural_features = ['Filopodia', 'Thin', 'Longthin', 'Mushroom', 'Stubby', 'Bifurcated', 'Critical', 'Non-critical', 'Total Spine Density (per 10 Âµm)']
melted_data = pd.melt(df, id_vars=['Developmental_Stage', 'Genotype'], value_vars=structural_features, var_name='Spine Type')

# Set the style for significance bars
def significance_annotate(p1, p2, text, h):
    if h == 0:
        h = 0.01
    y_max = max(p1.get_height(), p2.get_height())
    y = y_max + h
    x = p1.get_x() + (p2.get_x() - p1.get_x()) / 2
    plt.plot([p1.get_x(), p2.get_x()], [y, y], lw=1.5, c='black')
    plt.text(x, y, text, ha='center')

# Create a bar plot with significance bars
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=melted_data, x='Developmental_Stage', y='value', hue='Genotype', ci='sd', errwidth=1.5, capsize=0.15)
plt.xlabel('Developmental Stage')
plt.ylabel('Spine Density (per 10 Âµm)')
plt.title('Spine Density by Developmental Stage and Genotype')
plt.legend(title='Genotype')
plt.grid(True)

# Annotate significance bars
pubertal_wt = ax.patches[0]
pubertal_knockout = ax.patches[1]
postpubertal_wt = ax.patches[2]
postpubertal_knockout = ax.patches[3]

significance_annotate(pubertal_wt, pubertal_knockout, '*P < 0.05', h=0.02)
significance_annotate(postpubertal_wt, postpubertal_knockout, '*P < 0.05', h=0.02)

plt.tight_layout()
plt.show()
```

```python
# Calculate Critical and Noncritical spine densities
df['Critical'] = df['Mushroom'] + df['Stubby'] + df['Bifurcated']
df['Noncritical'] = df['Filopodia'] + df['Thin'] + df['Longthin']
df['Total'] = df['Critical'] + df['Noncritical']

# Interaction effect between Genotype and Developmental Stage
interaction_data = df.groupby(['Genotype', 'Developmental_Stage'])['Total'].mean().reset_index()
interaction_pivot = interaction_data.pivot('Genotype', 'Developmental_Stage', 'Total')

plt.figure(figsize=(8, 5))
sns.heatmap(interaction_pivot, annot=True, cmap='coolwarm', center=0)
plt.title('Interaction Effect between Genotype and Developmental Stage on Total Spine Densities')
plt.show()

# Effect of Genotype on Spine Types by Developmental Stage
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Genotype', y='Critical', hue='Developmental_Stage', ci='sd', capsize=0.1)
plt.title('Effect of Genotype on Critical Spine Densities by Developmental Stage')
plt.ylabel('Critical Spine Density')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Genotype', y='Noncritical', hue='Developmental_Stage', ci='sd', capsize=0.1)
plt.title('Effect of Genotype on Noncritical Spine Densities by Developmental Stage')
plt.ylabel('Noncritical Spine Density')
plt.tight_layout()
plt.show()



# Effect of Genotype on Total Spine Densities by Developmental Stage
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Developmental_Stage', y='Total', hue='Genotype', palette='Set2')
plt.title('Effect of Genotype on Total Spine Densities by Developmental Stage')
plt.ylabel('Total Spine Density')
plt.tight_layout()
plt.show()
```

```python
# Define spine colors and shapes
spine_colors = {'Wildtype': 'blue', 'Knockout': 'red'}
spine_shapes = {'Filopodia': 'o', 'Thin': 's', 'Longthin': 'D', 'Mushroom': '^', 'Stubby': 'v',
                'Bifurcated': 'p', 'Critical': '*', 'Non-critical': 'h', 'Total': '+'}

# Set up the plot
plt.figure(figsize=(10, 6))
sns.set_palette([spine_colors['Wildtype'], spine_colors['Knockout']])

# Create scatter plots for each spine type and genotype
for spine_type, shape in spine_shapes.items():
    for genotype, color in spine_colors.items():
        subset = df[(df['Genotype'] == genotype)]
        x_values = subset['PND'] + np.random.uniform(-0.2, 0.2, size=len(subset))  # Introduce randomness
        plt.scatter(x_values, subset[spine_type], label=f'{genotype} {spine_type}', marker=shape, color=color)

# Customize the plot
plt.xlabel('Postnatal Day (PND)')
plt.ylabel('Spine Density (per 10 Âµm)')
plt.title('Developmental Stage vs Spine Density by Spine Type and Genotype')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
